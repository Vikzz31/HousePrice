```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
df = sns.load_dataset('mpg')
df = df.dropna()
df = df.drop('name', axis=1)
df['origin'] = df['origin'].astype('category').cat.codes
X = df.drop('mpg', axis=1)
y = df['mpg']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
mse_scores = []
r2_scores = []
plt.figure(figsize=(15, 12))
fold = 1
all_y_test = []
all_y_pred = []
for train_idx, test_idx in kf.split(X_scaled):
    X_train, X_test = X_scaled[train_idx], X_scaled[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"\nFold {fold}")
    print(f"Mean Squared Error: {mse:.2f}")
    print(f"R² Score: {r2:.2f}")
    mse_scores.append(mse)
    r2_scores.append(r2)
    all_y_test.extend(y_test)
    all_y_pred.extend(y_pred)
    plt.subplot(3, 2, fold)
    plt.scatter(y_test, y_pred, alpha=0.7, edgecolors='b')
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # Perfect prediction line
    plt.xlabel('Actual MPG')
    plt.ylabel('Predicted MPG')
    plt.title(f'Fold {fold} Actual vs Predicted\nR² = {r2:.2f}')
    fold += 1
plt.tight_layout()
plt.show()
print("\nAverage Performance Across Folds:")
print(f"Average MSE: {np.mean(mse_scores):.2f}")
print(f"Average R² Score: {np.mean(r2_scores):.2f}")
plt.figure(figsize=(8, 6))
plt.scatter(all_y_test, all_y_pred, alpha=0.6, edgecolors='g')
plt.plot([min(all_y_test), max(all_y_test)], [min(all_y_test), max(all_y_test)], 'r--')
plt.xlabel('Actual MPG')
plt.ylabel('Predicted MPG')
plt.title('Overall Actual vs Predicted MPG (All folds combined)')
plt.show()
```
